```
import numpy as np
from IPython.display import Image

def NeuralNetwork(m1,m2,w1,w2,b):
  z = m1 * w1 + m2 * w2 + b
  return sigmoid(z)

def sigmoid(z):
  return 1/(1 + np.exp(-z))


w1 = np.random.rand()
w2 = np.random.rand()
b = np.random.rand()

Image('flexg.png')
```

```
pred_1 = NeuralNetwork(3, 1.5, w1, w2, b)
pred_2 = NeuralNetwork(2, 1, w1, w2, b)
pred_3 = NeuralNetwork(4, 1.5, w1, w2, b)
pred_4 = NeuralNetwork(3, 1, w1, w2, b)
pred_5 = NeuralNetwork(3.5, .5, w1, w2, b)

print(pred_1,pred_2,pred_3,pred_4,pred_5)
```
