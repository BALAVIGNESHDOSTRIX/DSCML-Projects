```
''' 
    Predict the If the Person can buy the Insurance Based on his age
'''

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('insurance.csv')
df.head()
```

```
plt.scatter(df.age,df.bought_insurance, marker='+', color='blue')
```

```
from sklearn.model_selection import train_test_split
X_train, x_test, Y_train, y_test = train_test_split(df[['age']], df.bought_insurance, test_size=0.1)
```

```
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,Y_train)
```

```
logmodel.predict(x_test)
```

```
x_test
```

```
logmodel.score(x_test,y_test)
```

```
logmodel.predict_proba(x_test)
```

```
person_age = int(input("Enter the Age: "))

def predictInsuranceIssue(age=None):
  return logmodel.predict([[age]])

result = predictInsuranceIssue(age=person_age)

if result is 1:
  print("He is to be bought Insurance")
else:
  print("He is not to be bought Insurance")

```
