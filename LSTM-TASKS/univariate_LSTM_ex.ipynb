```
X = X.reshape(X.shape[0], X.shape[1], 1)
X
```

```
# Create a LSTM Model

model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=(3,1), return_sequences=False))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
```

```
model.summary()
```

```
model.fit(X, y, epochs=600, verbose=1)
```

```
# Demonstrate the Prediction

x_input = np.array([50,60,70])
x_input = x_input.reshape((1,3,1))
# x_input
predict = model.predict(x_input, verbose=0)
predict
```
